## CREAET THE SCHEMA AND DELTA TABLES FOR BRONZE LAYER

```python
%sql
CREATE SCHEMA IF NOT EXISTS mycatalog.olist_ecommerce_bronze
```

```python
%sql
CREATE VOLUME IF NOT EXISTS mycatalog.olist_ecommerce_bronze.checkpoints
```

# CUSTOMERS

```python
%sql
CREATE TABLE IF NOT EXISTS mycatalog.olist_ecommerce_bronze.customers (
  customer_id STRING,
  customer_unique_id STRING,
  customer_zip_code_prefix INTEGER,
  customer_city STRING,
  customer_state STRING,
  source_file STRING,
  source_file_timestamp TIMESTAMP
)
USING DELTA
```

# GeoLocation

```python
%sql
CREATE TABLE IF NOT EXISTS mycatalog.olist_ecommerce_bronze.geolocation (
  geolocation_zip_code_prefix INTEGER,
  geolocation_lat DOUBLE,
  geolocation_lng DOUBLE,
  geolocation_city STRING,
  geolocation_state STRING,
  source_file STRING,
  source_file_timestamp TIMESTAMP
)
USING DELTA
```

# Order Items

```python
%sql
CREATE TABLE IF NOT EXISTS mycatalog.olist_ecommerce_bronze.order_items (
  order_id STRING,
  order_item_id INTEGER,
  product_id STRING,
  seller_id STRING,
  shipping_limit_date TIMESTAMP,
  price DOUBLE,
  freight_value DOUBLE,
  source_file STRING,
  source_file_timestamp TIMESTAMP
)
USING DELTA
```

# Order Payments

```python
%sql
CREATE TABLE IF NOT EXISTS mycatalog.olist_ecommerce_bronze.order_payments (
  order_id STRING,
  payment_sequential INTEGER,
  payment_type STRING,
  payment_installments INTEGER,
  payment_value DOUBLE,
  source_file STRING,
  source_file_timestamp TIMESTAMP
)
USING DELTA 
```

# Order Reviews

```python
%sql
CREATE TABLE IF NOT EXISTS mycatalog.olist_ecommerce_bronze.order_reviews (
  review_id STRING,
  order_id STRING,
  review_score INTEGER,
  review_comment_title STRING,
  review_comment_message STRING,
  review_creation_date TIMESTAMP,
  review_answer_timestamp TIMESTAMP,
  source_file STRING,
  source_file_timestamp TIMESTAMP
)
USING DELTA
```

# Orders

```python
%sql
    
CREATE TABLE IF NOT EXISTS mycatalog.olist_ecommerce_bronze.orders (
  order_id STRING,
  customer_id STRING,
  order_status STRING,
  order_purchase_timestamp TIMESTAMP,
  order_approved_at TIMESTAMP,
  order_delivered_carrier_date TIMESTAMP,
  order_delivered_customer_date TIMESTAMP,
  order_estimated_delivery_date TIMESTAMP,
  source_file STRING,
  source_file_timestamp TIMESTAMP
)
USING DELTA
```

# Product Categories

```python
%sql
CREATE TABLE IF NOT EXISTS mycatalog.olist_ecommerce_bronze.product_category (
  product_category_name STRING,
  product_category_name_english STRING,
  source_file STRING,
  source_file_timestamp TIMESTAMP
)
USING DELTA
```

# Products

```python
%sql
CREATE TABLE IF NOT EXISTS mycatalog.olist_ecommerce_bronze.products (
  product_id STRING,
  product_category_name STRING,
  product_name_lenght INTEGER,
  product_description_lenght INTEGER,
  product_photos_qty INTEGER,
  product_weight_g INTEGER,
  product_length_cm INTEGER,
  product_height_cm INTEGER,
  product_width_cm INTEGER,
  source_file STRING,
  source_file_timestamp TIMESTAMP
)
USING DELTA 
```

# Sellers

```python
%sql
   
CREATE TABLE IF NOT EXISTS mycatalog.olist_ecommerce_bronze.sellers (
  seller_id STRING,
  seller_zip_code_prefix INTEGER,
  seller_city STRING,
  seller_state STRING,
  source_file STRING,
  source_file_timestamp TIMESTAMP
)
USING DELTA
```
